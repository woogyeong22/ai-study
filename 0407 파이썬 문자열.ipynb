```python
license_plate = "24가 2210" # 문자열에서 여러글자를 가져오는 것을 슬라이싱이라부름
print(license_plate[-4:])   # 음수값은 문자열의 뒤에서부터 슬라이싱함을 의미
                            # 시작 인덱스를 생략하면 0으로 간주하고 끝 인덱스를 생략하면 문자열의 끝을 의미
                            # : <- 콜론이  슬라이싱의 시작과 끝을 구분함
```

```python
string = "홀짝홀짝홀짝"
print(string[::2])    # 첫번째문자부터 시작하여 두칸씩 건너뛰면서 문자를 추출하라는 의미
```

```python
string = "PYTHON"
print(string[::-1]) # 문자열을 뒤에서 부터 역순으로
```
