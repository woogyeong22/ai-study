```python
# 파이썬에서 True 혹은 False를 갖는 데이터 타입은 무엇인가?
# `bool` 타입이다.
```

```python
print(3 == 5) # Flase
```

```python
print(3 < 5) #True
```

```python
x = 4
print(1 < x < 5)  #True
```

```python
print(3 => 4)  #print(3 <= 4) oo
#  => 는 지원하지 않는 연산자
```

```python
if 4 < 3:
    print("Hello World")
    # 조건이 만족하지않아 출력 x
```

```python
if True :
    print ("1")
    print ("2")
else :
    print("3") # 조건이 True 이므로 건너뛰어짐
print("4") # if-else문과 별개의 문장이므로 조건과 관계 없이 항상 실행
```

```python
if True : # 가장 바깥 조건 : True
    if False: # 안쪽 조건 : False
        print("1")
        print("2")
    else:
        print("3")  # 안쪽 if 조건이 False였으므로 실행됨
else :
    print("4") # 바깥 if 조건이 True 였으므로 건너뛰어짐
print("5") #if-else문과 별개의 문장이므로 조건과 관계 없이 항상 실행
```
