```python
name1 = "김민수"
age1 = 10
name2 = "이철희"
age2 = 13
print("이름: %s 나이: %d" % (name1, age1))
print("이름: %s 나이: %d" % (name2, age2))
```

```python
name1 = "김민수"
age1 = 10
name2 = "이철희"
age2 = 13
print("이름: {} 나이: {}".format(name1, age1))    # format 함수는 가독성이 좋고 유연하게
print("이름: {} 나이: {}".format(name2, age2))    # 문자열을 만들 수 있어 % 연산자를 이용한
                                                  # 포맷팅보다 선호되는 방식이다.
```
